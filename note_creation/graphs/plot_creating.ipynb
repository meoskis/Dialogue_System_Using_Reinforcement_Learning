```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from matplotlib import rcParams
import pickle

rcParams['figure.figsize'] = (30, 20)
rcParams['figure.dpi'] = 300
rcParams['font.size'] = 20
rcParams['xtick.labelsize'] = 20
rcParams['ytick.labelsize'] = 20
rcParams['axes.titlesize'] = 24

rcParams['xtick.major.size'] = 12
rcParams['xtick.minor.size'] = 7
rcParams['xtick.major.width'] = 3
rcParams['xtick.minor.width'] = 1.5

rcParams['ytick.major.size'] = 12
rcParams['ytick.minor.size'] = 7
rcParams['ytick.major.width'] = 3
rcParams['ytick.minor.width'] = 1.5

rcParams['grid.color'] = 'DarkTurquoise'
rcParams['grid.linestyle'] = ':'
rcParams['grid.linewidth'] = 2

import warnings
warnings.filterwarnings("ignore")
```

```python
dqn_1 = pickle.load(open(r'C:\kudryashov_course_work\notebooks_and_program\dialog_run\data_for_graphs\DQN\success_rate_dict_1.pkl', 'rb'), encoding='latin1')
dqn_2 = pickle.load(open(r'C:\kudryashov_course_work\notebooks_and_program\dialog_run\data_for_graphs\DQN\success_rate_dict_2.pkl', 'rb'), encoding='latin1')
dqn_3 = pickle.load(open(r'C:\kudryashov_course_work\notebooks_and_program\dialog_run\data_for_graphs\DQN\success_rate_dict_3.pkl', 'rb'), encoding='latin1')
```

```python
dqn = pd.DataFrame(list(dqn_1.values()) + list(dqn_2.values()) + list(dqn_3.values()))
dqn = dqn[dqn!=0].dropna()
dqn.index = [idx*100 for idx in range(1,dqn.shape[0]+1)]
```

```python
x_dqn = np.asarray(dqn.index)
y_dqn = dqn[0].to_numpy()

x_max = []
y_max = []
maxi = 0
for succ in range(len(x_dqn)):
    if y_dqn[succ] > maxi:
        x_max.append(x_dqn[succ])
        y_max.append(y_dqn[succ])
        maxi = y_dqn[succ]

z = np.polyfit(x_dqn, y_dqn, 6)
p = np.poly1d(z)

plt.plot(x_dqn,y_dqn)
plt.plot(x_dqn,p(x_dqn))
plt.scatter(x_max,y_max);
```
