```python
res_user_goals = []
random.seed(100)

date_subset = ['сегодня','завтра','пятница']
time_subset = ['19:00', '20:00','18:30']
numofpeople_subset = ['2', '3', '4', '5']
numofkids_subset = ['2', '1', '3', '4', '5']

for i in range(110):
    rest_selection = random.choice(restaurants)
    indvl_user_goal = {'request_slots': {},
                       'diaact':'request',
                       'inform_slots':{
                       'город': rest_selection['город'],
                       'район': rest_selection['район'],
                       'название': rest_selection['название']   
                       }
                      }
    
    if random.random()<0.5:
        indvl_user_goal['inform_slots']['дата'] = random.choice(date_subset)
        if random.random()<0.5:
            indvl_user_goal['inform_slots']['время'] = random.choice(time_subset)
        else:
            indvl_user_goal['request_slots']['время'] = 'UNK'
    else:
        indvl_user_goal['request_slots']['дата'] = 'UNK'
        indvl_user_goal['request_slots']['время'] = 'UNK'
        
    
    if random.random()<0.65:
        indvl_user_goal['inform_slots']['количество_человек'] = random.choice(numofpeople_subset)
    else:
        indvl_user_goal['request_slots']['количество_человек'] = 'UNK'
        
    if random.random()<0.4:
        indvl_user_goal['inform_slots']['количество_детей'] = random.choice(numofpeople_subset)
    
    res_user_goals.append(indvl_user_goal)
```

```python
with open('C:/Users/nkmeo/Course work/Datasets/new_data/res_user_goals.txt', 'w') as filehandle:
    for listitem in res_user_goals:
        filehandle.write('%s\n' % listitem)
```
